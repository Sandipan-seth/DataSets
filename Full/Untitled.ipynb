```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
df=pd.read_csv("C:/Users/SANDIPAN SETH/Desktop/Full/Contact Information.csv")
df.rename(columns={'Enter your name': 'Name',
                  'Username': 'Email id',
                  'Enter Your Phone Number(optional)': 'Phone no',
                  'What is your gender?':'Gender',
                  'Enter your city':'City',
                  'Which category of cab you used most?':'Most use',
                  'What is your age':'Age',
                  'When you mostly use this service':'Time',
                  'Your average spending per ride':'Spending',
                  'What is your primary purpose ':'Purpose',
                  'How much fare do you expect to reserve a toto per km ride?':'Expectation',
                  'How much distance you want to cover with toto/auto?':'Distance'}, inplace=True)
df.replace('Never', np.nan, inplace=True)
df.replace('18-25', np.mean([18,25]), inplace=True)
df.replace('50-60', np.mean([50,60]), inplace=True)
df.replace('26-30', np.mean([26,30]), inplace=True)
df.replace('30-40', np.mean([30,40]), inplace=True)
df['City'] = df['City'].str.upper()
df['City'] = df['City'].str.replace(' ', '')
df.loc[df['City'] == 'DURGAPUR', 'City']="RISHRA"
df.loc[df['City'] == 'KONNOGOR', 'City']="KONNAGAR"
df.loc[df['City'] == 'SERAMPUR', 'City']="SERAMPORE"
df.loc[df['City'] == 'SHEORAPHULI', 'City']="SEORAPHULI"
print(list(df['Age']))
df.drop('Timestamp', axis=1, inplace=True)
selected_columns = df[['Name', 'Phone no', 'Email id','City']]
selected_columns.to_csv('information.csv', index=False)
df
```

```python
time_slots = ['8 am - 11 am', '12 pm - 5 pm', '6 pm - 9 pm', '9 pm - 12 am', 'In Night time']

# Filter and count the occurrences of the specific time slots
time_counts = df['Time'].value_counts().reindex(time_slots, fill_value=0)

# Plot the bar graph
plt.figure(figsize=(10, 5))
time_counts.plot(kind='bar', color='skyblue')

# Set the labels and title
plt.xlabel('Time')
plt.ylabel('Count')
plt.title('Serampore People Usage Time of Car Service')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=0, ha='center')

# Display the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
plt.figure(figsize=(5, 6))
plt.pie(time_counts, labels=time_counts.index, autopct='%1.1f%%', startangle=200, colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'])
plt.title('People Usage Time of Car Service')
plt.axis('equal') 
```

```python
desired_time_slots = ['8 am - 11 am', '12 pm - 5 pm', '6 pm - 9 pm', '9 pm - 12 am','In Night time']
df_filtered = df[df['Time'].isin(desired_time_slots)]

# Plotting the strip plot
plt.figure(figsize=(10, 6))
sns.stripplot(x='Time', y='Name', data=df_filtered, size=10, jitter=True, palette='Set2')

plt.title('Density of Car Service Usage in Serampore')
plt.xlabel('Time of Car Service')
plt.ylabel('Name')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
desired_time_stamps = ['8 am - 11 am', '12 pm - 5 pm', '6 pm - 9 pm', '9 pm - 12 am','In Night time']

# Filter the DataFrame to include only the desired time stamps
df_filtered = df[df['Time'].isin(desired_time_stamps)]

# Count occurrences of each time stamp
time_stamp_counts = df_filtered['Time'].value_counts().reindex(desired_time_stamps, fill_value=0)

# Plotting the line plot
plt.figure(figsize=(10, 6))
plt.plot(time_stamp_counts.index, time_stamp_counts.values, marker='o', linestyle='-', color='b', label='Count')
plt.title('Count of Specific Time Stamps')
plt.xlabel('Time Stamps')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
spending_categories = ['below 100', '100-300', '300-500', '500-800', '800-1000']

# Create a categorical dtype for 'spending' column to enforce order
spending_dtype = pd.api.types.CategoricalDtype(categories=spending_categories, ordered=True)
df['Spending'] = df['Spending'].astype(spending_dtype)

# Group by 'spending' and count occurrences of each 'Name'
spend_counts = df.groupby('Spending')['Name'].count()

# Plotting the point-line graph
plt.figure(figsize=(12, 6))
plt.plot(spend_counts.index, spend_counts.values, marker='+', linestyle=':', color='blue')

plt.title('Point-Line Graph of Spending vs Count of Name')
plt.xlabel('Spending')
plt.ylabel('Count of Name')
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.grid(True)  # Add gridlines
plt.tight_layout()
plt.show()
```
